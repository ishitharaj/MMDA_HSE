### Data Scaling 

### StandardScaler | Data Standardization

```python
'''
Standardizing a dataset involves rescaling the distribution of values so that the mean of observed
values is 0 and the standard deviation is 1.
This can be thought of as subtracting the mean value or centering the data.
Like normalization, standardization can be useful, and even required in some machine learning 
algorithms when your data has input values with differing scales.

A value is standardized as follows:
y = (x â€“ mean) / standard_deviation

Where the mean is calculated as:
mean = sum(x) / count(x)

And the standard_deviation is calculated as:
standard_deviation = sqrt(sum((x â€“ mean)^2 ) / count(x))

'''
```

```python
from sklearn.preprocessing import StandardScaler
Sscaler = StandardScaler()
```

```python
df_StScaled = Sscaler.fit_transform(df_new)
df_StScaled
```

### MinMaxScaler | Data Normalization

## Let's try k-means for random initializations | This is our homework task

```python
def best_solution(k):
    from math import inf # math module
    import random #random module
    
    trials = 10
    lowest_inertia = float(inf) #set to the positive infinity at the beginning
    best_intialized_state = None 
    best_centers = None 
    labels = None
    for i in range(trials):
        rand_state = random.randint(0,10000)
        kmeans = KMeans(n_clusters=k, init='random', random_state=rand_state)
        y_kmeans = kmeans.fit_predict(df_StScaled)
        if kmeans.inertia_ < lowest_inertia:
            lowest_inertia = kmeans.inertia_
            best_intialized_state = rand_state
            best_centers = kmeans.cluster_centers_
            labels = kmeans.labels_
            
    return {
        'inertia:':lowest_inertia,
        'best_centers': best_centers,
        'best_intialized_state':best_intialized_state,
        'labels': labels 
    }     
```

```python
best_solution5 = best_solution(5)
best_solution5
```

```python
best_solution9 = best_solution(9)
best_solution9
```

```python
df_new['K=5'] = best_solution5['labels']
df_new['K=9'] = best_solution9['labels']
```
